## Pyplot

```python
pyplot.plot(year,temp_anomaly)
```

```python
pyplot.rc('font', family='serif', size='18')  #坐标轴格式
```

```python
pyplot.plot(year, temp_anomaly, color ="green", linestyle='-', linewidth = 1)
```

```python
pyplot.figure(figsize=(10, 5))
pyplot.plot(year, temp_anomaly, color ="#2929a3", linestyle='-', linewidth = 1)
pyplot.title("lannd global temp anomalies. \n")
pyplot.xlabel("Year")
pyplot.ylabel("Land temperature anomaly [$\circ$C]")
pyplot.grid() #added; to avoid extra stuff appearing in image
# pyplot.savefig("myplot.png")
```

```python
a_1n, a_0n = np.polyfit(year, temp_anomaly, 1)
```

```python
print(a_1n)
```

```python
type(a_1n)
```

```python
f_linear = np.poly1d((a_1n, a_0n))
print(f_linear)
```

```python
pyplot.figure(figsize=(10, 5))
pyplot.plot(year, temp_anomaly, color ="blue", linestyle='-', linewidth = 1)
pyplot.plot(year, f_linear(year), "k--", linewidth=2, label = 'Linear regression')
pyplot.title("lannd global temp anomalies. \n")
pyplot.xlabel("Year")
pyplot.ylabel("Land temperature anomaly [$\circ$C]")
pyplot.legend(loc = 'best', fontsize = 15)
pyplot.grid() #added; to avoid extra stuff appearing in image
# pyplot.savefig("myplot.png")
```

```python
year_1, temp_anomaly_1 = year[0:90], temp_anomaly[0:90]
year_2, temp_anomaly_2 = year[90:], temp_anomaly[90:]
```

```python
m1, b1 = np.polyfit(year_1, temp_anomaly_1, 1)
m2, b2 = np.polyfit(year_2, temp_anomaly_2, 1)
```

```python
print(m1,b1)
```

```python
f_linear_1 = np.poly1d((m1, b1))
f_linear_2 = np.poly1d((m2, b2))

print(f_linear_1)
print(f_linear_1)
```

```python
pyplot.figure(figsize=(10, 5))
pyplot.plot(year, temp_anomaly, color ="blue", linestyle='-', linewidth = 1)
pyplot.plot(year_1, f_linear_1(year_1), "g--", linewidth=2, label = '1880-1969')
pyplot.plot(year_2, f_linear_2(year_2), "r--", linewidth=2, label = '1969-2017')
pyplot.title("lannd global temp anomalies. \n")
pyplot.xlabel("Year")
pyplot.ylabel("Land temperature anomaly [$\circ$C]")
pyplot.legend(loc = 'best', fontsize = 15)
pyplot.grid() #added; to avoid extra stuff appearing in image
# pyplot.savefig("myplot.png")
```
